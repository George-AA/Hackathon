# Processing the Big Five Personality Model

After reading on the usual output for this personality test, we thought to map our answers to a specific percentage for each of the 5 personality traits.

We have a set of discrete answers to the questions, making the mapping easier.

We associate each answer to a numerical value which we will use to later output each person's scoring on the Big Five Traits.

We map the answers as follows:


*   Agree Strongly => +1 value
*   Agree moderately => +0.5
*   Agree a Little => +0.25
*   Neither Agree nor Disagree => 0
*   Disagree a Little => -0.25
*   Disagree Moderately => -0.5
*   Disagree Strongly => -1

We mapped the questions as follows:

* Extraversion
  *  "I see myself as someone who is extraverted, enthusiastic."
  *  "I see myself as someone who is reserved, quiet." (reversely mapped)
* Agreeableness
  * "I see myself as someone who is sympathetic, warm."
  * "I see myself as someone who is critical, quarrelsome." (reversely mapped)
* Conscientiousness
  * "I see myself as someone who is dependable, self-disciplined."
  * "I see myself as someone who is disorganized, careless." (reversely mapped)
* Neuroticism
  * "I see myself as someone who is anxious, easily upset."
  * "I see myself as someone who is calm, emotionally stable." (reversely mapped)
* Openness
  * "I see myself as someone who is open to new experiences."
  * "I see myself as someone who is conventional, uncreative." (reversely mapped)


```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd


response_mapping = {
    "Agree strongly": 1,
    "Agree moderately": 0.5,
    "Agree a little": 0.25,
    "Neither agree nor disagree": 0,
    "Disagree a little": -0.25,
    "Disagree moderately": -0.5,
    "Disagree strongly": -1
}

def map_answer(answer):
    return response_mapping.get(answer, 0)


try:
    data = pd.read_excel('2024_PersonalityTraits_SurveyData.xls', sheet_name='Sheet1')
except FileNotFoundError:
    print("Error: 'your_excel_file.xlsx' not found. Please upload the file or provide the correct path.")
    exit()

extraversion_cols = ["I see myself as someone who is extraverted, enthusiastic:", "I see myself as someone who is reserved, quiet:"]
agreeableness_cols = ["I see myself as someone who is sympathetic, warm:", "I see myself as someone who is critical, quarrelsome:"]
conscientiousness_cols = ["I see myself as someone who is dependable, self-disciplined:", "I see myself as someone who is disorganized, careless:"]
neuroticism_cols = ["I see myself as someone who is anxious, easily upset:", "I see myself as someone who is calm, emotionally stable:"]
openness_cols = ["I see myself as someone who is open to new experiences:", "I see myself as someone who is conventional, uncreative:"]


reverse_cols = {
    "I see myself as someone who is reserved, quiet:": True,
    "I see myself as someone who is critical, quarrelsome:": True,
    "I see myself as someone who is disorganized, careless:": True,
    "I see myself as someone who is calm, emotionally stable:": True,
    "I see myself as someone who is conventional, uncreative:": True
}

def calculate_trait_score(df, cols):
    scores = []
    for index, row in df.iterrows():
        trait_score = 0
        for col in cols:
            answer = row[col]
            score = map_answer(answer)
            if col in reverse_cols and reverse_cols[col]:
                score *= -1
            trait_score += score
        scores.append(trait_score)
    return scores

extraversion_scores = calculate_trait_score(data, extraversion_cols)
agreeableness_scores = calculate_trait_score(data, agreeableness_cols)
conscientiousness_scores = calculate_trait_score(data, conscientiousness_cols)
neuroticism_scores = calculate_trait_score(data, neuroticism_cols)
openness_scores = calculate_trait_score(data, openness_cols)

trait_scores_matrix = pd.DataFrame({
    "Extraversion": extraversion_scores,
    "Agreeableness": agreeableness_scores,
    "Conscientiousness": conscientiousness_scores,
    "Neuroticism": neuroticism_scores,
    "Openness": openness_scores
})

normalized_scores = 100 * (trait_scores_matrix - trait_scores_matrix.min()) / (trait_scores_matrix.max() - trait_scores_matrix.min())

print(normalized_scores)
```

# Processing Smoking Dependency
For this section of the data, we encoded different attributes relating to one's dependency on nicotine, turning them from discrete values to numerical ones and normalizing them.

```python
frequency_mapping = {
    "10 or less cigarettes/day": 1,
    "11 to 20 cigarettes": 2,
    "21 to 30 cigarettes":3,
    "31 cigarettes/day or more": 4
}

def map_frequency(answer):
    return frequency_mapping.get(answer, 0)/4


def map_yesorno(answer):
    if answer=="Yes":
        return 1
    else:
        return 0

first_cigarette = {
    "Within 5 minutes": 4,
    "6 to 30 minutes": 3,
    "31 to 60 minutes": 2,
    "After 60 minutes": 1
}

def map_first_cigarette(answer):
    return first_cigarette.get(answer, 0)

def map_cigarette_to_drop(answer):
    if answer=="The first one in the morning":
        return 1
    elif answer=="All others":
        return 0


def map_age(answer):
  answer = int(answer)
  x = None
  if answer<=12:
    x = 3
  elif answer<=17:
    x = 2
  elif answer<=25:
    x = 1
  else:
    x = 0
  return x/3

def map_rate(answer):
  x = None
  if answer == "The number of cigarettes I smoke per day has remained the same":
    x = 2
  elif answer == "The number of cigarettes I smoke per day has increased":
    x = 3
  elif answer == "The number of cigarettes I smoke per day has decreased":
    x = 1
  else:
    x = 0
  return x/3

smoke_when_forbidden_col = data['Do you find it difficult to refrain from smoking where it is forbidden (church, library, cinema, plane, etc...)?']
freq_col = data['How many cigarettes do you smoke each day?']
time_col = data['Do you smoke more frequently during the first hours after waking up than during the rest of the day?\xa0']
smoke_when_sick_col = data['Do you smoke if you are so ill that you are in bed most of the day?']
first_cigarette_col = data['How soon after you wake up do you smoke your first cigarette?']
cigarette_to_drop_col = data['Which cigarette would you mostly hate to give up?']
age_of_smoke_col = data['How old were you the first time you smoked a full cigarette (not just a few puffs)?']
change_habit_col = data["How would you describe your current smoking behavior compared to your smoking behavior before Lebanon's economic crisis and revolution began in 2019?"]

def calculate_dep_score(col, x):
    scores = []
    for row in col:
        trait_score = 0
        answer = row
        score = x(answer)
        trait_score += score
        scores.append(trait_score)
    return scores

smoke_when_forbidden_col_scores = calculate_dep_score(smoke_when_forbidden_col, map_yesorno)
freq_col_scores = calculate_dep_score(freq_col, map_frequency)
time_col_scores = calculate_dep_score(time_col, map_yesorno)
smoke_when_sick_col_scores = calculate_dep_score(smoke_when_sick_col, map_yesorno)
first_cigarette_col_scores = calculate_dep_score(first_cigarette_col, map_first_cigarette)
cigarette_to_drop_col_scores = calculate_dep_score(cigarette_to_drop_col, map_cigarette_to_drop)
age_of_smoke_col_scores = calculate_dep_score(age_of_smoke_col, map_age)
change_habit_col_scores = calculate_dep_score(change_habit_col, map_rate)

dependence_matrix = pd.DataFrame({
    "Smoking When Forbidden": smoke_when_forbidden_col_scores,
    "Cigarettes Per Day": freq_col_scores,
    "Smoking When Sick": smoke_when_sick_col_scores,
    "Smoking More in the First Hours": time_col_scores,
    "First Cigarette": first_cigarette_col_scores,
    "Cigarette to Drop": cigarette_to_drop_col_scores,
    "Age of Smoke First Full Cigarette": age_of_smoke_col_scores,
    "Change Smoking Habit": change_habit_col_scores
})

normalized_dep_scores = 100 * (dependence_matrix - dependence_matrix.min()) / (dependence_matrix.max() - dependence_matrix.min())

print(normalized_dep_scores)
```

# Processing Demographic Data
For this section of the data, we encoded different attributes relating to one's demographic, turning them from discrete values to numerical ones and normalizing them.

```python
def gender(answer):
  if answer=="Male":
    return 1
  else:
    return 0

def age(answer):
  return (int(answer))

governates=["Beirut","Nabatieh","Mount Lebanon","North Lebanon","Keserwan - Jbeil","Beqaa","South Lebanon","Baalbeck - Hermel"]

def governate(answer):
  return governates.index(answer)

def marital_status(answer):
  if answer=="Married":
    return 3
  elif answer=="Engaged":
    return 2
  elif answer=="Single":
    return 1
  else:
    return 0

def smoker_friends(answer):
  return int(answer)

source_of_incomes=["Parents","Own business income","Investment","Job","Other, please specify"]
def source_of_income(answer):
  return source_of_incomes.index(answer)

def income(answer):
  if answer=="Fully in Lebanese Lira":
     return 2
  elif answer=="Fully in US Dollars":
     return 3
  elif answer=="Mixed":
     return 1
  else:
     return 0

def monthly_income(answer):
  if answer=="More than 20 million L.L":
    return 7
  elif answer=="I don't know":
    return 1
  elif answer=="Between 1 and 4 million L.L":
    return 2
  elif answer=="Between 12 and 16 million L.L":
    return 5
  elif answer=="Between 8 and 12 million L.L":
    return 4
  elif answer=="Between 4 and 8 million L.L":
    return 3
  elif answer=="Between 16 and 20 million L.L":
    return 6
  else:
    return 0

def income_sufficiency(answer):
  if answer=="High: completely covers necessities with a few luxury items":
     return 4
  elif answer=="Medium: covers all basic needs":
     return 3
  elif answer=="Low: barely covers basic needs for a month":
     return 2
  elif answer=="Very low income: does not cover basic needs for a month":
     return 1
  elif answer=="Extremely high: covers a wide range of luxury items":
     return 5
  else:
     return 0

def education(answer):
   if answer=="Graduate degree (MA/MS)":
       return 5
   elif answer=="Bachelor's degree (BA/BS)":
       return 4
   elif answer=="Incomplete bachelor's degree":
       return 3
   elif answer=="Post-graduate degree (PhD, MD, or other)":
       return 6
   elif answer=="High school degree or equivalent (e.g. GED)":
       return 2
   elif answer=="Less than high school":
       return 1
   else:
       return 0

def people_in_household(answer):
  return int(answer)

def economic_effect(answer):
  if answer=="Very":
    return 4
  elif answer=="Slightly":
    return 3
  elif answer=="Moderately":
    return 2
  elif answer=="Not at all":
    return 1
  elif answer=="Extremely":
    return 5
  else:
    return 0

def exercise(answer):
  if answer=="Never":
    return 4
  elif answer=="Often or at least 3 days every week":
    return 3
  elif answer=="Every day or at least 5 times a week":
    return 2
  elif answer=="Sometimes or a few days every month":
    return 1
  elif answer=="Extremely":
    return 5
  else:
    return 0

def social_media(answer):
  if answer=="Less than 1 hour":
    return 5
  elif answer=="Between 1 and 2 hours":
    return 4
  elif answer=="Between 2 and 3 hours":
    return 3
  elif answer=="Between 3 and 4 hours":
    return 2
  elif answer=="More than 4 hours":
    return 1
  else:
    return 0

def stressed(answer):
  if answer=="Frequently":
    return 4
  elif answer=="Occasionally":
    return 3
  elif answer=="Constantly":
    return 5
  elif answer=="Never":
    return 1
  elif answer=="Rarely":
    return 2

def employment(answer):
  if answer=="Employed":
    return 1
  else:
    return 0

def employment_status(answer):
  if answer=="Business owner":
    return 3
  elif answer=="Student only":
    return 1
  elif answer=="Student with a part-time or full-time job":
    return 2
  else:
    return 0

gender_col = data['Gender:']
age_col = data['How old are you?']
governate_col = data['Which governerate do you live in or spend most of your time in?']
education_col = data["What is the highest level of education you have attained?"]
employment_status_col = data['What is your current employment status?']
marital_status_col = data['What is your current marital status?']
close_friends_col = data['Do you have close friends?']
smoker_friends_col = data["Of the five closest friends or acquaintances that you spend time with on a regular basis, how many of them are smokers?"]
source_of_income_col = data['What is your main source of income?']
income_col = data['What type of income or financial support does your household receive?']
income_amount_col = data["If you receive payment in Lebanese Lira, what is your current estimated monthly household income? (If income is in US Dollars, then refer to the current black market exchange)."]
income_sufficiency_col = data['How would you describe your current income sufficiency?']
people_in_household_col = data['Including yourself, how many people currently live in your household?']
economic_effect_col = data["To what extent were you financially (negatively) affected by the deterioration of the Lebanese economy?"]
exercise_col = data["How often do you exercise?"]
social_media_col = data["On average, how many hours per day do you spend on social media for entertainment and social interaction (Facebook, Instagram, YouTube, etc...)?"]
stress_col = data['How often do you feel stressed?']
employment_status2_col = data["Employment Status"]


gender_col_scores = calculate_dep_score(gender_col, gender)
age_col_scores = calculate_dep_score(age_col, age)
governate_col_scores = calculate_dep_score(governate_col, governate)
education_col_scores = calculate_dep_score(education_col, education)
employment_status_col_scores = calculate_dep_score(employment_status_col, employment_status)
marital_status_col_scores = calculate_dep_score(marital_status_col, marital_status)
close_friends_col_scores = calculate_dep_score(close_friends_col, map_yesorno)  # Assuming Yes/No format
smoker_friends_col_scores = calculate_dep_score(smoker_friends_col, smoker_friends)
source_of_income_col_scores = calculate_dep_score(source_of_income_col, source_of_income)
income_col_scores = calculate_dep_score(income_col, income)
income_amount_col_scores = calculate_dep_score(income_amount_col, monthly_income)
income_sufficiency_col_scores = calculate_dep_score(income_sufficiency_col, income_sufficiency)
people_in_household_col_scores = calculate_dep_score(people_in_household_col, people_in_household)
economic_effect_col_scores = calculate_dep_score(economic_effect_col, economic_effect)
exercise_col_scores = calculate_dep_score(exercise_col, exercise)
social_media_col_scores = calculate_dep_score(social_media_col, social_media)
stress_col_scores = calculate_dep_score(stress_col, stressed)
employment_status2_col_scores = calculate_dep_score(employment_status2_col, employment)

demographic_socioeconomic_matrix = pd.DataFrame({
    "Gender": gender_col_scores,
    "Age": age_col_scores,
    "Governate": governate_col_scores,
    "Education Level": education_col_scores,
    "Employment Status": employment_status_col_scores,
    "Marital Status": marital_status_col_scores,
    "Close Friends": close_friends_col_scores,
    "Smoker Friends": smoker_friends_col_scores,
    "Source of Income": source_of_income_col_scores,
    "Income Type": income_col_scores,
    "Monthly Income": income_amount_col_scores,
    "Income Sufficiency": income_sufficiency_col_scores,
    "People in Household": people_in_household_col_scores,
    "Economic Effect": economic_effect_col_scores,
    "Exercise Frequency": exercise_col_scores,
    "Social Media Usage": social_media_col_scores,
    "Stress Levels": stress_col_scores,
    "Employment": employment_status2_col_scores
})

normalized_demographic_socioeconomic_scores = 100 * (
    demographic_socioeconomic_matrix - demographic_socioeconomic_matrix.min()
) / (demographic_socioeconomic_matrix.max() - demographic_socioeconomic_matrix.min())

print(normalized_demographic_socioeconomic_scores)
```

# Clustering
Finally, after encoding all the data, we use a gaussian mixture model to cluster our data.

```python
from sklearn.mixture import GaussianMixture

all_data = pd.concat([normalized_scores, normalized_dep_scores, normalized_demographic_socioeconomic_scores], axis=1)

gmm = GaussianMixture(n_components=3, random_state=0)

gmm.fit(all_data)

labels = gmm.predict(all_data)

all_data['cluster'] = labels

all_data
```

```python
import matplotlib.pyplot as plt

cluster_centers = gmm.means_

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', label='Cluster Centers')

plt.scatter(all_data.iloc[:, 0], all_data.iloc[:, 1], c=all_data['cluster'], cmap='viridis', label='Data Points')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Gaussian Mixture Model Clusters')
plt.legend()

plt.show()
```
